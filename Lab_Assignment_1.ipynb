```python
import pandas as pd

# Define the file path
file_path = r"C:\Users\DELL\Downloads\archive (1)\Dataset_Generator_for_DTDC.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
import matplotlib.pyplot as plt

# Scatter plot to show relationship between 'Tariff' and 'Total Amount'
plt.figure(figsize=(10, 6))
plt.scatter(df['Tariff'], df['Total Amount'])
plt.title('Scatter Plot: Tariff vs Total Amount')
plt.xlabel('Tariff')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Bar chart for 'Mode of Payment'
df['Mode of Payment'].value_counts().plot(kind='bar', figsize=(10, 6))
plt.title('Bar Chart: Mode of Payment')
plt.xlabel('Mode of Payment')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for 'Tariff'
df['Tariff'].hist(bins=30, figsize=(10, 6))
plt.title('Histogram: Distribution of Tariff')
plt.xlabel('Tariff')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for 'Tariff'
plt.figure(figsize=(10, 6))
df.boxplot(column='Tariff')
plt.title('Box Plot: Tariff')
plt.show()
```

```python
# Pie chart for 'Nature of Consignment'
df['Nature of Consignment'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Pie Chart: Nature of Consignment')
plt.ylabel('')
plt.show()
```

```python
# Check for missing values in the 'Tariff' column
df['Tariff'].isnull().sum()


```

```python
df['Tariff'] = df['Tariff'].fillna(df['Tariff'].mean())
```

```python
# Check the data type of the 'Tariff' column
df['Tariff'].dtype
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Scale the 'Tariff' column
df['Tariff_scaled'] = scaler.fit_transform(df[['Tariff']])

# Display the dataset with scaled 'Tariff'
df.head()
```

```python
# General info and summary statistics
df.info()
df.describe()
```

```python
# Calculate central tendency measures for 'Tariff'
min_value = df['Tariff'].min()
max_value = df['Tariff'].max()
mean_value = df['Tariff'].mean()
median_value = df['Tariff'].median()
mode_value = df['Tariff'].mode()[0]

print(f"Min: {min_value}, Max: {max_value}, Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}")
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'Tariff'
Q1 = df['Tariff'].quantile(0.25)
Q3 = df['Tariff'].quantile(0.75)

# Calculate the IQR (Interquartile Range)
IQR_value = Q3 - Q1

# Now you can calculate other measures like Range, Variance, etc.
min_value = df['Tariff'].min()
max_value = df['Tariff'].max()
range_value = max_value - min_value
variance = df['Tariff'].var()
std_dev = df['Tariff'].std()

# Display the results
print(f"Range: {range_value}, Variance: {variance}, Standard Deviation: {std_dev}, IQR: {IQR_value}")

```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix on the numeric columns
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print(correlation_matrix)

```
